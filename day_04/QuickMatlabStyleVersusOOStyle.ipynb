```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2

print x[:3]
print y[:3]
```

### Matlab Style Interface (Quick and Dirty - Exploratory Stuff)

What goes on with matlab-style interface:  
  * [MPL FAQ](http://matplotlib.org/faq/usage_faq.html#matplotlib-pyplot-and-pylab-how-are-they-related)
  * and the FAQ point on "coding style" below that

```python
plt.plot(x, y, 'r')
plt.xlabel('Dollars')
plt.ylabel('Burgers')
plt.title('Burgers for Dollars in my Wallet')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

### Let's Get Explicit (better than Implicit):  The OO-Interface

```python
fig = plt.figure()

graph = fig.add_axes([0, 0, 1, 0.3]) # left, bottom, width, height (range 0 to 1)

graph.plot(x, y, 'r')

graph.set_xlabel('x')
graph.set_ylabel('y')
graph.set_title('title');
```

```python
fig = plt.figure()

graph1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
graph2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
graph1.plot(x, y, 'r')
graph1.set_xlabel('Dollars')
graph1.set_ylabel('Burgers')
graph1.set_title('Made a Plot with OO Interface\n')

# inset 
# (for example, zooming in on an important region of a graph)
graph2.plot(x, y, 'g')
graph2.set_xlabel('Dollars')
graph2.set_ylabel('Burgers')
graph2.set_title('Burgers near Zero');
```

```python
fig, axes = plt.subplots(figsize=(15,5))
# -or-
# fig.set_figheight fig.set_figwidth 
# fig.set_size_inches (for both)
axes.plot(x, y, 'r')
axes.set_xlabel('Dollars')
axes.set_ylabel('Burgers');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True)

styles = ['bo-','r+']
for style, ax in zip(styles, axes):
    ax.plot(x, y, style)
    ax.set_xlabel('Dollars')
    ax.set_ylabel('Burgers')
    ax.set_title('Burgers versus Dollars')
```
